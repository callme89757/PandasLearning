```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Liquid, Page,Gauge,WordCloud,Line
from pyecharts.globals import SymbolType
import tushare as ts
import math as Math
from pyecharts.commons import utils
```

```python
aFrame=ts.get_stock_basics()
npr_mean=aFrame['npr'].mean() #平均净利率
gpr_mean=aFrame['gpr'].mean()/100 #平均毛利率
c=(
    Liquid()
    .add('lq',
        [gpr_mean],
        label_opts=opts.LabelOpts(
            font_size=50,
            formatter=utils.JsCode(
                """function (param) {
                    return (Math.floor(param.value * 10000) / 100) + '%';
                }"""
            ),
                position="inside",
        ),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="A股市场平均毛利率"))
)
c.render_notebook()
```

```python
g=(
Gauge()
.add("",[("净利率",round(npr_mean,2))],min_=-100)
.set_global_opts(title_opts=opts.TitleOpts(title="A股平均净利率图"))
)
g.render_notebook()
```

```python
info_company=ts.get_industry_classified().groupby('c_name')['name'].agg([('公司数量','count')])
info_company_name=info_company.index.values.tolist()
info_company_num=info_company['公司数量'].values.tolist()
words=zip(info_company_name,info_company_num)
wordPhoto=(
    WordCloud()
    .add("", words, word_size_range=[20, 100])
    .set_global_opts(title_opts=opts.TitleOpts(title="A股行业词云图"))
)
wordPhoto.render_notebook()
```

```python
df=ts.get_hist_data('600519')
high=df['high'].values.tolist()
high.reverse()
low=df['low'].values.tolist()
low.reverse()
xname=df.index.values.tolist()
xname.reverse()
li=(
    Line()
    .add_xaxis(xname)
    .add_yaxis("最高价",high,markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="max")]))
    .add_yaxis("最低价", low,markline_opts=opts.MarkLineOpts(data=[opts.MarkLineItem(type_="min")]))
    .set_global_opts(title_opts=opts.TitleOpts(title="茅台近2年的日内最高价与最低价折线图"),
                    yaxis_opts=opts.AxisOpts(min_=450,),
                    datazoom_opts = opts.DataZoomOpts()
                    )
)
li.render_notebook()
```
