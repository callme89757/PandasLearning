```python
import pandas as pd
from matplotlib import pyplot as plt
f=open("D:/python/myprogram/PandaLearning/task_2_lianjia_data.csv",encoding='utf-8')
lianjia_data=pd.read_table(f,sep=',')
df1=pd.DataFrame(lianjia_data)
df1.head(8)

df2=df1.groupby('面积大小（㎡）')["简介（方式·小区名 户型 朝向）"].agg({"房源数量":'count'})
```

```python
lst_num=df2['房源数量'].tolist()
lst_area=df2.index.values.tolist()
plt.figure(figsize=(20,10))
plt.bar(lst_area,lst_num, alpha=0.9, width = 0.8, facecolor = 'lightskyblue', edgecolor = 'white', lw=1)
plt.plot(lst_area,lst_num,'bo--')
#结论1：上海房源面积以25~50平的小户型居多，最小的户型为7平米，最大的户型为173平米
```
