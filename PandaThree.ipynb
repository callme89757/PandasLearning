```python
import pandas as pd
from matplotlib import pyplot as plt
f=open("D:/python/myprogram/PandaLearning/task_3.1_zhihu_timeline_answer.csv",encoding='utf-8')
df=pd.read_table(f,sep=',')
zhihu_data_answer=pd.DataFrame(df)
zhihu_data_answer.head()
```

```python
lst_bingo_num=zhihu_data_answer['回答点赞数'].tolist()
lst_thank_num=zhihu_data_answer['回答感谢数'].tolist()
```

```python
ax1=plt.figure(figsize=(20,10)).add_subplot(111)
ax1.set_title("点赞与感谢散点图")
plt.rcParams['font.sans-serif']=['SimHei']
plt.xlabel("回答点赞")
plt.ylabel("回答感谢")
ax1.scatter(lst_bingo_num,lst_thank_num,c='r',marker='o')
```

```python
#结论1：绝大部分回答的点赞数<5000，感谢数<2000，极个别回答获得了5000以上的感谢。

lst_answer_num=zhihu_data_answer['问题回答数'].tolist()
lst_focue_num=zhihu_data_answer['问题关注数'].tolist()
ax2=plt.figure(figsize=(20,10)).add_subplot(111)
ax2.set_title("回答与关注散点图")
plt.xlabel("问题回答数")
plt.ylabel("问题关注数")
ax2.scatter(lst_answer_num,lst_focue_num,c='b')
```

```python
#结论2：几乎所有的回答数都在2000以内，大部分问题关注数小于5000，但也有一部分问题关注数在5000~30000之间不均匀分布
```

```python
f.close()
f=open("D:/python/myprogram/PandaLearning/task_3.2_zhihu_timeline_article.csv",encoding='utf-8')
df1=pd.read_table(f,sep=',')
zhihu_data_article=pd.DataFrame(df1)
zhihu_data_article.head()
```

```python
lst_bingo_num=zhihu_data_article['文章点赞数'].tolist()
lst_discuss_num=zhihu_data_article['文章评论数'].tolist()
ax3=plt.figure(figsize=(20,10)).add_subplot(111)
ax3.set_title("点赞与评论散点图")
plt.xlabel("点赞数")
plt.ylabel("评论数")
ax3.scatter(lst_bingo_num,lst_discuss_num,c='y')
```

```python
#结论3：绝大部分文章评论数在1000以内，点赞数在10000以内，评论数超过4000的只有一篇文章
```

```python
lst_bingo_num1000=[]
for l in lst_bingo_num:
    if l<1000:
        lst_bingo_num1000.append(l)
plt.boxplot(lst_bingo_num1000,showmeans=1)
```
