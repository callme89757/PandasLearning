```python
np.random.seed(1)
values = np.random.randn(1000).cumsum()
from matplotlib import pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(range(0,1000),values,'bo--')
```

```python
max_drawdown = np.max(np.maximum.accumulate(values) - values)
max_drawdown
```
